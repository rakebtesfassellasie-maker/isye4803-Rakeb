```python
import pandas as pd
import statsmodels.formula.api as smf
```

```python
roe = pd.read_csv("../roe01.txt", sep=r"\s+")

roe.head()
```

```python
print("Column means:")
print(roe.mean())

print("\nCorrelation between ROEt and ROE:")
print(roe["ROEt"].corr(roe["ROE"]))

model = smf.ols("ROE ~ ROEt", data=roe).fit()

print("\nLinear regression summary:")
print(model.summary())
```

If my Python and R results disagreed, one possible reason could be missing values, since NA values are common in financial data. One language might drop rows with missing values, while the other might keep them or handle them differently. A second reason could be a different modeling decision, such as using different variables, selecting different rows, or making an indexing mistake because Python starts at 0 while R starts at 1. Also, using  different defaults, such as population variance versus sample variance. Population variance divides by n however the sample variance divides by n - 1, so the result may not match. To decide which result is correct, I would check the original dataset, confirm that both languages are using the same rows and columns, and make sure the same formula and missing-value rules are being used.
